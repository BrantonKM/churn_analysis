##### ONE-HOT ENCODING CATEGORICAL FEATURES.


To be able to run a classification model categorical features are transformed into dummy variable values of 0 and 1

```python
# 'international_plan' is the column to encode:
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(df2[['international plan']])

# Create a DataFrame from the encoded data:
encoded_df2 = pd.DataFrame(encoded_data, columns=['international_plan_yes', 'international_plan_no'])

# Concatenate the encoded DataFrame with the original DataFrame:
df2 = pd.concat([df2, encoded_df2], axis=1)

# Drop the original column if needed:
df2.drop('international plan', axis=1, inplace=True)
```

```python
# 'voice mail plan' is the column to encode:
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(df2[['voice mail plan']])

# Create a DataFrame from the encoded data:
encoded_df2 = pd.DataFrame(encoded_data, columns=['voice_mail_plan_yes', 'voice_mail_plan_no'])

# Concatenate the encoded DataFrame with the original DataFrame:
df2 = pd.concat([df2, encoded_df2], axis=1)

# Drop the original column if needed:
df2.drop('voice mail plan', axis=1, inplace=True)
```

```python
# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Encode the 'area code' column
encoded_data = encoder.fit_transform(df2[['area code']])

# Create a DataFrame from the encoded data
encoded_df2 = pd.DataFrame(encoded_data, columns=['area_code_' + str(int(code)) for code in encoder.categories_[0]])

# Concatenate the encoded DataFrame with the original DataFrame
df2 = pd.concat([df2, encoded_df2], axis=1)

# Drop the original 'area code' column if needed
df2.drop('area code', axis=1, inplace=True)
```
